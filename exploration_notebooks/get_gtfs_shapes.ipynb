```python
import geopandas as gpd
```

```python
tm_routes = gpd.read_file('gis/tm_routes.shp')
tm_routes.head(3)
```

```python
tm_routes.crs
```

```python
tm_routes_prj = tm_routes.to_crs("EPSG:4326")
```

```python
tm_routes_prj['rte_dir'] = tm_routes_prj.apply(lambda x: str(x['rte'])+'_'+str(x['dir']), axis=1)
```

```python
tm_routes_prj.head(3)
```

```python
tm_routes_prj.to_file('gis/export/tm_routes.geojson', driver="GeoJSON")
```
